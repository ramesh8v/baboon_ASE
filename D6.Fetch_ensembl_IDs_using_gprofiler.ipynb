```python
import pandas as pd
```

```python
df = pd.read_csv("../../rsem/baboon_genome/panu.gtf", sep="\t", comment="#", names = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute'])
```

```python
df.head(2)
```

```python
df = df[df['feature'] == 'gene']
```

```python
df['gene_id'] = df['attribute'].str.extract('"GeneID:(.+?)";')
```

```python
genes = df['gene_id'].to_list()
```

```python
from gprofiler import GProfiler

gp = GProfiler(return_dataframe=True)
df2 = gp.convert(organism='panubis',
           query=genes,
           target_namespace='ENSG',
            numeric_namespace='ENTREZGENE_ACC')
```

```python
df2
```

```python
df2[df2['converted'] == "None"].shape, df2.shape
```

```python
df1 = df.merge(df2, left_on='gene_id', right_on='incoming')
```

```python
df1
```

```python
df1['biotype'] = df1['attribute'].str.extract('gene_biotype "(.+?)";')
```

```python
df1[['seqname', 'source', 'feature', 'start', 'end', 'attribute', 'gene_id', 'converted', 'name', 'biotype']]
```

```python
df1.to_csv("data_files/ncbi_plus_ensebl_IDs.tsv", sep="\t", index=None)
```
